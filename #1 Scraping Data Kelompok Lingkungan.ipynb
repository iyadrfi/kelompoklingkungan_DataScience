# Data Bank Sampah di BPS Sultra

**BANYAKNYA DESA/KELURAHAN MENURUT KEBERADAAN BANK SAMPAH DAN PENGOLAHAN SAMPAH

```python
alamat = "https://sultra.bps.go.id/statictable/2022/09/28/4215/banyaknya-desa-kelurahan-menurut-keberadaan-bank-sampah-dan-pengolahan-sampah.html"
req = Request(alamat, headers={'User-Agent': 'Mozilla/5.0'})
```

```python
html = urlopen(req).read()
data = BeautifulSoup(html, 'html.parser')
div_isi_tabel = data.find('div', {'id': 'isi-tabel', 'class': 'tabel'})
table = div_isi_tabel.find_all('table')[0]
```

```python
rows = table.findAll('tr')
```

```python
hasil = []
for row in rows:
   cols = row.find_all('td')
   cols = [ele.text.strip() for ele in cols]
   hasil.append(cols)
```

```python
# Megambil data dari baris 5
df_hasil = pd.DataFrame(hasil[5:])  
df_hasil = df_hasil.reset_index(drop=True)  

# Menghapus baris 22-24
df_hasil = df_hasil.drop(df_hasil.index[17:22])  
df_hasil = df_hasil.reset_index(drop=True)  

# Menghapus kolom 5 sampai 7
df_hasil = df_hasil.drop(df_hasil.columns[3:8], axis=1)
df_hasil
```

```python
# memberi judul kolom tabel
df_hasil.columns = ['Kabupaten/Kota', 'Bank Sampah Ada', 'Bank Sampah TA'] 
```

```python
df_hasil
```

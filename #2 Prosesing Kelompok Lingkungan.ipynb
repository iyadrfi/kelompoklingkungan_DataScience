# Data Preprocessing

```python
# Konversi kolom ke tipe data numerik 
data["Ketersediaan Tempat Pembuangan Sampah Sementara"] = pd.to_numeric(data["Ketersediaan Tempat Pembuangan Sampah Sementara"], errors='coerce')
data["Pengolahan/Daur Ulang Sampah/Limbah Ada"] = pd.to_numeric(data["Pengolahan/Daur Ulang Sampah/Limbah Ada"], errors='coerce')

data["Sumber Air Minum Bersih"] = data["Sumber Air Minum Bersih"].apply(lambda x: x.replace(',', '.') if isinstance(x, str) else x)
data["Sumber Air Minum Layak"] = data["Sumber Air Minum Layak"].apply(lambda x: x.replace(',', '.') if isinstance(x, str) else x)

data["Sumber Air Minum Bersih"] = pd.to_numeric(data["Sumber Air Minum Bersih"], errors='coerce')
data["Sumber Air Minum Layak"] = pd.to_numeric(data["Sumber Air Minum Layak"], errors='coerce')


A = data["Ketersediaan Tempat Pembuangan Sampah Sementara"]
B = data["Pengolahan/Daur Ulang Sampah/Limbah Ada"]
C = data["Sumber Air Minum Bersih"]
D = data["Sumber Air Minum Layak"]

data["Indeks Kesehatan"] = (0.30 * A + 0.20 * B + 0.25 * C + 0.25 * D) 

data["Indeks Kesehatan"] = data["Indeks Kesehatan"].round(1)

indeks_kesehatan = data[["Kabupaten/Kota", "Indeks Kesehatan"]]
print(indeks_kesehatan)
```

```python
indeks_terbesar = indeks_kesehatan.sort_values(by="Indeks Kesehatan", ascending=False)
indeks_terbesar.reset_index(drop=True, inplace=True)
indeks_terbesar.index = indeks_terbesar.index + 1
print(indeks_terbesar)
```

Kita telah mendapatkan nilai Indeks Kesehatannya dengan rumus :
Faktor A (Ketersediaan Tempat Pembuangan Sampah Sementara) diberikan bobot 30%,
Faktor B (Pengolahan/Daur Ulang Sampah/Limbah Ada) diberikan bobot 20%,
Faktor C (Sumber Air Minum Bersih) diberikan bobot 25%,
Faktor D (Sumber Air Minum Layak) diberikan bobot 25%.

## Visualisasi Dalam Peta 

```python
data = pd.DataFrame({
    'Kabupaten/Kota': ['Kendari', 'Bau-bau', 'Kolaka', 'Wakatobi', 'Bombana', 'Buton Tengah', 'Buton', 'Kolaka Utara', 'Buton Utara', 'Konawe', 'Konawe Selatan', 'Muna', 'Kolaka Timur', 'Konawe Utara', 'Muna Barat', 'Konawe Kepulauan', 'Buton Selatan'],
    'Indeks Kesehatan': [65.9, 60.3, 58.0, 55.1, 52.5, 49.2, 48.3, 48.0, 47.9, 47.6, 47.5, 46.8, 46.1, 44.8, 41.8, 38.1, 37.0],
    'Latitude': [-3.9984597, -5.507078, -3.9946988, -5.67046665, -4.6543462, -5.291780999999999, -5.3096355, -3.1347227, -4.702342400000001, -3.9380432, -4.2027915, -4.901629, -4.2279225, -3.3803291, -4.7990321, -4.1361465, -5.3096355],
    'Longitude': [122.5129742, 122.596901, 121.5826642, 123.79394565, 121.9017954, 122.3787597, 122.9888319, 121.1710389, 123.0338767, 122.0837445, 122.4467238, 122.6277455, 121.9017954, 122.0837445, 122.4969606, 123.1239049, 122.9888319]
})

# Buat peta dengan folium
m = folium.Map(location=[-3.484772, 122.247140], zoom_start=8)

# Urutkan data indeks kesehatan
data = data.sort_values(by='Indeks Kesehatan', ascending=False)

# Tambahkan marker dengan ikon angka dan teks peringkat pada peta
for index, row in data.iterrows():
    rank = index + 1  
    
    
    if rank <= 5:
        background_color = 'red'  
    elif rank <= 10:
        background_color = 'blue'  
    else:
        background_color = 'green'  
    
    # Buat marker dengan ikon angka dan teks peringkat
    marker = folium.Marker(
        location=[row['Latitude'], row['Longitude']],
        icon=DivIcon(
            icon_size=(20, 20),
            icon_anchor=(10, 10),
            html=f'<div style="font-size: 12px; background-color: {background_color}; color: white; border-radius: 50%; border: 1px solid #000; width: 20px; height: 20px; text-align: center; line-height: 20px;">{rank}</div>'
        ),
    )
    
    # Tambahkan popup teks peringkat
    popup_text = f'Kabupaten/Kota: {row["Kabupaten/Kota"]}<br>Peringkat: {rank}<br>Indeks Kesehatan: {row["Indeks Kesehatan"]}'
    folium.Popup(popup_text, max_width=200).add_to(marker)
    
    # Tambahkan marker ke peta
    marker.add_to(m)

# Tampilkan peta
m.save('peta_peringkat_indeks_kesehatan.html')
```
